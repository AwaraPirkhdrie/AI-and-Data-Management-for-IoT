A.0.1_Day_Old_Bread 

```python
'''
This Python script solves the problems outlined in assignment A1.1, sub exercises I and IV.
Author: Awara Pirkhdrie
Date: 2024-01-18

A bakery sells loaves of bread for $3.49 each. Day-old bread is discounted by 60 percent.
Write a program that begins by reading the number of loaves of day-old bread being purchased from the user.
Then your program should display the regular price for the bread, the discount because it is a day old,
and the total price. All of the values should be displayed using two decimal places, and the decimal points
in all of the numbers should be aligned when reasonable values are entered by the user.
'''

def calculate_bread_costs():

    '''
    The price for each loaf of bread, Discount rate for day-old bread (60%)
    '''
    _BREAD_PRICE_FOR = 3.49
    _DAY_OLD_BREAD_DISCOUNTED = 0.60

    '''
    Prompting user to enter the number of loaves and converting it to an integer.
    Check if the input number is negative, which is not allowed.
    Handles cases where the input is not an integer or is negative.
    Exit the function if the input is invalid.
    '''
    try:
        _number_of_loaves_os_day_old_bread = int(input("Enter the number of day-old loaves: "))
        if _number_of_loaves_os_day_old_bread < 0:
            raise ValueError("Number of loaves cannot be negative.")
    except ValueError as e:
        print("Invalid input. Please enter a whole number.")
        return

    '''
     Calculating the total regular price for the entered number of loaves.
     Calculating the amount of discount based on the discount rate.
     Calculating the final price after applying the discount.
     Printing the calculated prices, formatted to two decimal places.
     '''
    _stander_price = _number_of_loaves_os_day_old_bread * _BREAD_PRICE_FOR
    _bread_is_discount = _stander_price * _DAY_OLD_BREAD_DISCOUNTED
    _total_price = _stander_price - _bread_is_discount
    print(f"\nRegular price (for {_number_of_loaves_os_day_old_bread} loaves): ${_stander_price:.2f}")
    print(f"Discount: ${_bread_is_discount:.2f}")
    print(f"Total price after discount: ${_total_price:.2f}")

calculate_bread_costs()
```

A.0.2_Dog_Years

```python
'''

This Python script solves the problems outlined in assignment A1.1, sub exercises I and IV.
Author: Awara Pirkhdrie
Date: 2024-01-18


It is commonly said that one human year is equivalent to 7 dog years. However,
this simple conversion fails to recognize that dogs reach adulthood in approximately two years.
As a result, some people believe that it is better to count each of the first two human years as 10.5 dog years,
 and then count each additional human year as 4 dog years.
Write a program that implements the conversion from human years to dog years described in the previous paragraph.
Ensure that your program works correctly for conversions of less than two human years and for conversions
of two or more human years. Your program should display an appropriate error message if the user enters a negative number.
'''

'''
The function human_years_to_dog_years converts human years to dog years. The user inputs human years,
and the function uses a two-part formula: 10.5 dog years per human year up to two years,
then 4 dog years per additional human year. Invalid entries are handled with error messages.
'''
def human_years_to_dog_years():
    try:
        # User input
        _number_of_human_years = float(input("Enter the number of human years: "))

        # Validate input
        if _number_of_human_years < 0:
            print("Invalid input. Number of years cannot be negative.")
            return

        # Conversion calculation
        if _number_of_human_years <= 2:
            # For 2 or fewer human years, each year counts as 10.5 dog years.
            _number_of_dog_years = _number_of_human_years * 10.5
        else:
            # For more than 2 human years, the first two count as 21 dog years (2 * 10.5),
            # and each additional year counts as 4 dog years.
            _number_of_dog_years = 21 + (_number_of_human_years - 2) * 4

        # Display result
        print(f"{_number_of_human_years} human years are equivalent to {_number_of_dog_years} dog years.")

    except ValueError:
        # Handle cases where the input is not a number
        print("Invalid input. Please enter a valid number.")

# Call the function
human_years_to_dog_years()
```

A.0.3_Coin_Flip_3_same_flips

```python
'''

This Python script solves the problems outlined in assignment A1.1, sub exercises I and IV.
Author: Awara Pirkhdrie
Date: 2024-01-18


Create a program that uses Python’s random number generator to simulate flipping a coin several times.
The simulated coin should be fair, meaning that the probability of heads is equal to the probability of tails.

Each time the program is run, your program should flip simulated coins until either
 3 consecutive heads or 3 consecutive tails occur. Display an H each time the outcome is heads,
 and a T each time the outcome is tails, with all of the outcomes shown on the same line.
 Then display the number of flips needed to reach 3 consecutive flips with the same outcome.
'''

import random

'''
Simulates flipping a coin with equal probability of 'H' (head) and 'T' (tail). Returns the result of the flip.
'''
def flip_coin():
    return 'H' if random.random() < 0.5 else 'T'

'''
Tracks the number of consecutive matches (heads or tails) when flipping a coin, and counts the total number of flips until three matches are consecutive.
'''
def main():
    _consecutive_count = 0
    _last_flip = None
    _flip_sequence = ''
    _flip_count = 0

    while _consecutive_count < 3:
        _current_flip = flip_coin()
        _flip_sequence += _current_flip
        _flip_count += 1

        # Check if the current flip is the same as the last flip
        if _current_flip == _last_flip:
            _consecutive_count += 1
        else:
            _consecutive_count = 1
            _last_flip = _current_flip

    print(_flip_sequence)
    print(f"Number of flips needed: {_flip_count}")

'''
Makes the main function run when the script is executed directly, not when imported as a module.
'''
if __name__ == "__main__":
    main()


```

A.0.4_Is _it_a_prime_number

```python
'''

This Python script solves the problems outlined in assignment A1.1, sub exercises I and IV.
Author: Awara Pirkhdrie
Date: 2024-01-18


A prime number is an integer greater than 1 that is only divisible by one and itself.
Write a function that determines whether or not its parameter is prime, returning True if it is,
and False otherwise. Write a main program that reads an integer from the user and displays
a message indicating whether or not it is prime. Ensure that the main program will not run
if the file containing your solution is imported into another program.
'''

'''
Determines whether a number is a prime number. Returns False for numbers ≤ 1 and numbers divisible by some smaller number.
'''
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

'''
The main program that asks the user to enter an integer, uses is_prime to check if the number is prime, and displays the result.
'''
def main():
    try:
        _number = int(input("Enter an integer: "))
        if is_prime(_number):
            print(f"{_number} is a prime number.")
        else:
            print(f"{_number} is not a prime number.")
    except ValueError:
        print("Invalid input. Please enter an integer.")

'''
Checks that the code runs as a main program and not as an import. If true, execute the main function.
'''
if __name__ == "__main__":
  main()

```

A.0.5_What_is_the_next_prime

```python
'''

This Python script solves the problems outlined in assignment A1.1, sub exercises I and IV.
Author: Awara Pirkhdrie
Date: 2024-01-18


 In this exercise, you will create a function named "nextPrime" that finds and returns
 the first prime number larger than some integer, n. The value of n will be passed to the
 function as its only parameter. Include a main program that reads an integer from the user and
 displays the first prime number larger than the entered value. Import and use your solution to
 Exercise A.03 while completing this exercise.
'''

'''
Checks whether an integer is a prime number. Returns False for numbers less than or equal to 1, and for all non-prime numbers.
'''
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

'''
Finds the first prime number greater than the given number n. Increments successively from n + 1 until a prime number is found.
'''
def nextPrime(n):
    # Start with the next number after n
    _number_larger_than_some_integer = n + 1

    # Loop until a prime number is found
    while True:
        if is_prime(_number_larger_than_some_integer):
            return _number_larger_than_some_integer
        _number_larger_than_some_integer += 1

'''
The main program takes an input of an integer from the user, calculates the next prime after this number, and displays the result.
'''
def main():
    try:
        _number_input = int(input("Enter an integer: "))
        _next_prime = nextPrime(_number_input)
        print(f"The first prime number larger than {_number_input} is {_next_prime}.")
    except ValueError:
        print("Invalid input. Please enter an integer.")

'''
Ensures that the main function is executed when the script is executed directly and not when it is imported as a module.
'''
if __name__ == "__main__":
    main()

```

A.0.6_Shuffle_the_cards

```python
'''

This Python script solves the problems outlined in assignment A1.1, sub exercises I and IV.
Author: Awara Pirkhdrie
Date: 2024-01-19


A standard deck of playing cards contains 52 cards. Each card has one of four suits along with a value.
The suits are normally spades, hearts, diamonds and clubs while the values are
2 through 10, Jack, Queen, King and Ace. Each playing card can be represented using two characters.
The first character is the value of the card, with the values 2 through 9 being represented directly.
The characters “T”, “J”, “Q”, “K” and “A” are used to represent the values 10, Jack, Queen, King and Ace
respectively. The second character is used to represent the suit of the card. It is normally a lowercase
letter: “s” for spades, “h” for hearts, “d” for diamonds and “c” for clubs. The following table provides
several examples of cards and their two-character representations.

 Begin by writing a function named createDeck. It will use loops to create a complete deck of cards by
 adding the suit to the number and storing the two-character abbreviations for all 52 cards in a list. Return
  the list of cards as the function’s only result. Your function will not take any parameters.

 Write a second function named shuffle that randomizes the order of the cards in a list.
 The input of the function should be the list contacting the deck of cards (52 items)
 that you have created in the previous function. You can use the "random" library shuffle function.
 Alternatively, you can write your own shuffling function:  One technique that can be used to shuffle
 the cards is to visit each element in the list and swap it with another random element in the list.
  You must write your own loop for shuffling the cards. The output of the function should be 2 lists:
  the deck of cards before shuffling and the deck of cards after shuffling.
'''



import random

class CardDeck:
    '''
    Initializes the CardDeck class. Creates a list of playing card suits and values, and combines these to create a deck.
    '''
    def __init__(self):
        self.suits = ['s', 'h', 'd', 'c']  # spades, hearts, diamonds, clubs
        self.values = ['2', '3', '4', '5', '6', '7', '8', '9', 'T', 'J', 'Q', 'K', 'A']
        self.deck = [value + suit for suit in self.suits for value in self.values]
    '''
    Shuffles the deck in a random order by using random.shuffle on the deck list.
    '''
    def shuffle(self):
        random.shuffle(self.deck)
    '''
    Displays the deck divided into four rows, each row representing a suit, by printing 13 cards at a time.
    '''
    def display(self):
        for i in range(0, len(self.deck), 13):
            print(' '.join(self.deck[i:i+13]))
'''
The main function that creates a deck, displays it, accepts user input to decide whether to shuffle the deck, and handles errors.
'''
def main():
    _deck = CardDeck()
    print("Original Deck:")
    _deck.display()

    try:
        _user_input = input("Do you want to shuffle the deck? (yes/no): ").strip().lower()
        if _user_input not in ["yes", "no"]:
            raise ValueError("Please enter 'yes' or 'no'.")

        if _user_input == "yes":
            _deck.shuffle()
            print("\nShuffled Deck:")
            _deck.display()
        else:
            print("\nDeck remains unshuffled.")
    except Exception as e:
        print(f"Error: {e}")
'''
Checks if the script is run directly and not imported, if so the main function is run.
'''
if __name__ == "__main__":
    main()
```

A.0.7_Deal_the_cards

```python
'''
This Python script solves the problems outlined in assignment A1.1, sub exercises I and IV.
Author: Awara Pirkhdrie
Date: 2024-01-22

 In many card games, each player is dealt a specific number of cards after the deck has been shuffled. Write a function, deal, which takes the number of hands,
 the number of cards per hand, and a deck of cards as its three parameters. Your function should return a list containing all of the hands that were dealt.
 Each hand will be represented as a list of cards. When dealing with the hands, your function should modify the deck of cards passed to it as a parameter,
 removing each card from the deck as it is added to a player’s hand. When cards are dealt, it is customary to give each player a card before any player receives an additional card.
 Your function should follow this custom when constructing the hands for the players.

When done with writing the above function, use your solution to A.0.6 to help you construct a main class that creates,
shuffles, and deals a deck of cards. Your class structure should be like below:
'''
import random

class Cards:
    '''
    Initialize a new instance of the Cards class. Creates an empty list of cards and calls the create method.
    self represents a class instance and is required to access any variables or methods within the class
    '''
    def __init__(self):
        self.cards = []
        self.create()
    '''
    Creates a standard deck of 52 cards. Uses list comprehension to combine each value with each color.
    '''
    def create(self):
        suits = ['s', 'h', 'd', 'c']
        values = ['2', '3', '4', '5', '6', '7', '8', '9', 'T', 'J', 'Q', 'K', 'A']
        self.cards = [value + suit for suit in suits for value in values]
    '''
    Shuffles the deck in a random order. Uses random.shuffle to reorder the elements in the self.cards list.
    '''
    def shuffle(self):
        random.shuffle(self.cards) # random blandning
    '''
    Deals cards to a specified number of hands. Checks if there are enough cards, deals cards sequentially.
    Not enough cards in deck to deal
    '''
    def deal(self, hands, card_num):
        if hands * card_num > len(self.cards):
            raise ValueError("Not enough cards in the deck to deal.")

        dealt_hands = [[] for _ in range(hands)]
        for i in range(card_num):
            for hand in dealt_hands:
                if self.cards:
                    hand.append(self.cards.pop(0)) # take the first place in the map
                else:
                    break
        return dealt_hands
'''The main function that initiates the game. Handles user input for number of players and cards per hand, creates and deals cards.'''
def main():
    try:
        card_01 = Cards()
        card_01.shuffle()
        X = int(input("Enter the number of hands: ")) # hur många personer är ni?
        print("People: ", X)
        Y = int(input("Enter the number of cards per hand: ")) # hur många kart cill ni ha
        print("Hands: ", Y)
        # X = 5 # välja 5 spelare
        # Y= 10 # välja 10 kort
        dealt_hands = card_01.deal(X, Y)
        for index, hand in enumerate(dealt_hands): # en inbyggd funktion som låter dig iterera genom en sekvens och hålla reda på indexet för varje element
            print(f"Hand {index+1}: {' '.join(hand)}")
        print(dealt_hands)
    except ValueError as ve: # error meddelande
        print(f"Error: {ve}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}") # error meddelande
''' Checks if the script is run directly (and not imported), and then runs the main function.'''
if __name__ == "__main__":
    main()
```
